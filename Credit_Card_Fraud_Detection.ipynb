## Spliting the data into X and y train/test

```python
X = data.drop('Class', axis = 1)
y = data['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test , y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 42)
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
```

```python
classifier = {
    'Logistic Regression' : LogisticRegression(),
    'Decision Tree' : DecisionTreeClassifier()
}
for name, clf in classifier.items():
    print(f"\n=========={name}===========")
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(f"\n Accuracy: {accuracy_score(y_test, y_pred)}")
    print(f"\n Precision: {precision_score(y_test, y_pred)}")
    print(f"\n Recall: {recall_score(y_test, y_pred)}")
    print(f"\n F1 Score: {f1_score(y_test, y_pred)}")
```

```python
# model is imbalanced, we can just check the accuracy, not the precision, nor another score.
```

## **To fix this issue, we will treat the data imbalance**

## **Under Sampling** - remove data till the 0 and 1 are equal size.
Issue: lost data*

```python
normal = data[data['Class']== 0 ]
fraud = data[data['Class']== 1 ]
```

```python
normal.shape
```

```python
fraud.shape
```

```python
normal_sample = normal.sample(n=473)
```

```python
normal_sample.shape
```

```python
new_data = pd.concat([normal_sample, fraud], ignore_index = True)
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
X = new_data.drop('Class', axis = 1)
y = new_data['Class']
```

```python
X_train, X_test , y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 42)
```

```python
classifier = {
    'Logistic Regression' : LogisticRegression(),
    'Decision Tree' : DecisionTreeClassifier()
}
for name, clf in classifier.items():
    print(f"\n=========={name}===========")
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(f"\n Accuracy: {accuracy_score(y_test, y_pred)}")
    print(f"\n Precision: {precision_score(y_test, y_pred)}")
    print(f"\n Recall: {recall_score(y_test, y_pred)}")
    print(f"\n F1 Score: {f1_score(y_test, y_pred)}")
```

## We can clearly see, how the scores improved after balancing the data.

## **Over Sampling**




```python
X = data.drop('Class', axis = 1)
y = data['Class']
```

```python
X.shape
```

```python
y.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X_res, y_res =  SMOTE().fit_resample(X, y)
```

```python
y_res.value_counts()
```

```python
X_train, X_test , y_train, y_test = train_test_split(X_res, y_res, test_size = .2, random_state = 42)
```

```python
classifier = {
    'Logistic Regression' : LogisticRegression(),
    'Decision Tree' : DecisionTreeClassifier()
}
for name, clf in classifier.items():
    print(f"\n=========={name}===========")
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(f"\n Accuracy: {accuracy_score(y_test, y_pred)}")
    print(f"\n Precision: {precision_score(y_test, y_pred)}")
    print(f"\n Recall: {recall_score(y_test, y_pred)}")
    print(f"\n F1 Score: {f1_score(y_test, y_pred)}")
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_res, y_res)
```
